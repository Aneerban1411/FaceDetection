```python
#importing matplotlib for plotting
import matplotlib.pyplot as plt
#importing mtcnn model from mtcnn module
from mtcnn.mtcnn import MTCNN
#importing opencv
import cv2
```

```python
filename = "/Users/aneerbanchakraborty/Documents/work/ml/DataSets-master/facial_recognition/image1.jpg"
#reading image
pixels = cv2.imread(filename)
pixels=cv2.cvtColor(pixels, cv2.COLOR_BGR2RGB)
```

```python
#initialising model
model= MTCNN()
```

```python
#detecting faces
faces= model.detect_faces(pixels)
```

```python
#printing faces
print(len(faces))
```

```python
#plotting all faces
for face in faces:
    x,y,w,h = face["box"]
    cv2.rectangle(pixels, (x,y), (x+w,y+h), (0,0,255), 2)
    lefteye=face["keypoints"]["left_eye"]
    righteye=face["keypoints"]["right_eye"]
    nossy=face["keypoints"]["nose"]
    leftmouth=face["keypoints"]["mouth_left"]
    rightmouth=face["keypoints"]["mouth_right"]
    cv2.circle(pixels, lefteye, 1, (0,0,255), 1)
    cv2.circle(pixels, righteye, 1, (0,0,255), 1)
    cv2.circle(pixels, nossy, 1, (0,0,255), 1)
    cv2.circle(pixels, leftmouth, 1, (0,0,255), 1)
    cv2.circle(pixels, rightmouth, 1, (0,0,255), 1)
    
```

```python
plt.imshow(pixels)
plt.show()
```

```python
#extracting each face
for i in faces:
    x,y,w,h = i["box"]
    face=pixels[y:y+h, x:x+w]
    plt.imshow(face)
    plt.show()
```
